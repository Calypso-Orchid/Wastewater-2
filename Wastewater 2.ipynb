```python
import pandas as pd
treatment = pd.read_csv('wastewater_edited.csv')
treatment = treatment.sort_values(by='% Wastewater Treated', ascending=False)
print(treatment)
```

```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
%matplotlib inline
```

```python
x = treatment['Country or Area']
x
```

```python
y = treatment['% Wastewater Treated']
y
```

```python
plt.barh(x, y)
plt.xlabel('% Wastewater Treated')
plt.ylabel('Country')
plt.title('Figure 1: Country vs. % Wastewater Treated')
plt.show()
```

```python
#This graph shows how percentage of wastewater treatment varies amongst countries. 
#El Salvador has the lowest percentage of population connected at 2.83 percent.
#Singapore and Kuwait have 100 percent of their population connected to wastewater treated.
```

```python
x1 = treatment['Continent']
x1
```

```python
y1 = treatment['% Wastewater Treated']
y1
```

```python
colors = treatment['% Wastewater Treated']
area = np.pi * (treatment['% Wastewater Treated'])
plt.scatter(x1, y1, s=area, c=colors, alpha=0.5)
plt.xlabel('Continent')
plt.ylabel('% Wastewater Treated')
plt.title('Figure 2: Continent vs. % Wastewater Treated')
plt.show()
```

```python
#This graph explores a different variable to see how percentage of wastewater treatment varies amongst continents. 
#It's interesting to note that treatment varies greatly in Europe, with some countries having very high treatment rates in northern Europe
#and lower treatment percentages in Eastern Europe.
```
